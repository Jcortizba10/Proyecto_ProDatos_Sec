```python
import re
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import chardet
import pandas as pd
# Validador de texto

class TextValidator:
    def validate(self, text):
        if len(text.strip()) < 3:
            return False
        if not re.match(r"^[a-zA-Z0-9ñÑáéíóúÁÉÍÓÚüÜ\s.,!?-]+$", text):  # Se permiten acentos y guiones
            return False
        if text.strip().isdigit():
            return False
        return True

# Modelo de Hugging Face
class HuggingFaceModel:
    def __init__(self, model_name="jcortizba/modelo18", validator=None):
        self.model_name = model_name
        self.tokenizer = AutoTokenizer.from_pretrained(self.model_name)
        self.model = AutoModelForSequenceClassification.from_pretrained(self.model_name)
        self.label_mapping = {0: "ejecutar", 1: "cancelar"}
        self.validator = validator or TextValidator()

    def predict(self, text):
        if not self.validator.validate(text):
            return "Entrada inválida. Por favor, ingrese un texto coherente."

        inputs = self.tokenizer(text, return_tensors="pt")
        outputs = self.model(**inputs)
        probabilities = torch.nn.functional.softmax(outputs.logits, dim=-1)
        predicted_index = torch.argmax(probabilities, dim=1).item()
        return self.label_mapping[predicted_index]

# Clasificación del texto transcrito
hf_model = HuggingFaceModel()
classification_result = hf_model.predict(result['text'])
print("Resultado de la clasificación:", classification_result)
```

```python
# Función para integrar Whisper y el modelo de Hugging Face
def transcribe_and_classify(audio_path):
    result = model.transcribe(audio_path)
    transcribed_text = result['text']
    hf_model = HuggingFaceModel()
    classification_result = hf_model.predict(transcribed_text)
    return transcribed_text, classification_result
```

```python
import gradio as gr

# Interfaz de Gradio
demo = gr.Interface(
    fn=transcribe_and_classify,
    inputs=gr.Audio(type="filepath"),
    outputs=[gr.Textbox(label="Transcripción del Audio"), gr.Textbox(label="Resultado de la Clasificación")],
    title="Transcripción y Clasificación de Audio",
    description="Sube un archivo de audio para transcribirlo usando Whisper y clasificar el texto usando un modelo de Hugging Face."
)

demo.launch()
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
from ydata_profiling import ProfileReport
from streamlit_ydata_profiling import st_profile_report
import whisper
from transformers import pipeline
import moviepy.editor as mp
import tempfile
import os
import warnings

warnings.filterwarnings("ignore", category=FutureWarning, module="whisper")

# Configuración de la página
st.set_page_config(page_title="EDA con Pandas Profiling", layout="wide")

# Título de la aplicación
st.title("Proyecto con el uso de Whisper")

# Cargar el modelo Whisper
model = whisper.load_model("base")

# Función para extraer el audio del video
def extract_audio(video_file, output_audio_file="temp_audio.wav"):
    video = mp.VideoFileClip(video_file)
    video.audio.write_audiofile(output_audio_file)
    return output_audio_file

# Función para transcribir el audio con timestamps
def transcribe_audio(audio_file):
    result = model.transcribe(audio_file, word_timestamps=True)
    return result['text']

# Función para generar un resumen
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

def summarize_text(text):
    summary = summarizer(text, max_length=130, min_length=30, do_sample=False)
    return summary[0]['summary_text']

# Función para análisis de sentimiento
sentiment_analyzer = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")

def analyze_sentiment(text):
    return sentiment_analyzer(text)

# Interfaz de usuario con Streamlit
def main():
    # Colocar el cargador de archivos en la barra lateral
    video_file = st.sidebar.file_uploader("Sube un archivo de video", type=["mp4", "mkv", "avi"])

    if video_file:
        st.write("Procesando el video...")

        # Crear un archivo temporal para el video
        with tempfile.NamedTemporaryFile(delete=False, suffix=".mp4") as temp_video_file:
            temp_video_file.write(video_file.read())
            temp_video_file.close()

        # Extraer audio
        st.write("Extrayendo audio del video...")
        audio_file = extract_audio(temp_video_file.name)

        # Transcribir audio con timestamps
        st.write("Transcribiendo el audio...")
        transcription = transcribe_audio(audio_file)
        st.write("**Texto transcrito:**")
        st.write(transcription)

        # Generar resumen
        st.write("Generando resumen del texto...")
        summary = summarize_text(transcription)
        st.write("**Resumen:**")
        st.write(summary)

        # Analizar sentimiento
        st.write("Analizando sentimiento...")
        sentiment = analyze_sentiment(transcription)
        st.write("**Sentimiento detectado:**")
        st.write(sentiment)

        # Limpiar archivos temporales
        os.remove(temp_video_file.name)
        os.remove(audio_file)

if __name__ == "__main__":
    main()
```
